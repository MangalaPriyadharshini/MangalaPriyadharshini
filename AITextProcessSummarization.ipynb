```python
# Install dependencies
!pip install -q langchain langchain-huggingface faiss-cpu sentence-transformers transformers langchain-community

from langchain_huggingface import HuggingFaceEmbeddings, HuggingFacePipeline
from langchain_community.vectorstores import FAISS  # <-- note change
from langchain.document_loaders import TextLoader
from langchain.chains import RetrievalQA
from transformers import pipeline

# 1️⃣ Load your document
loader = TextLoader("/content/AI_Full_Course_Overview.txt")
docs = loader.load()

# 2️⃣ Create embeddings + FAISS vector store
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vector_db = FAISS.from_documents(docs, embeddings)

# 3️⃣ HuggingFace causal LM pipeline
hf_pipeline = pipeline(
    "text-generation",
    model="bigscience/bloom-560m",  # causal LM, works on CPU
    max_new_tokens=150
)
llm = HuggingFacePipeline(pipeline=hf_pipeline)

# 4️⃣ Create RetrievalQA chain
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=vector_db.as_retriever(),
    chain_type="stuff"
)

query = "What is AI?"
response = qa_chain.invoke({"query": query})
print(response['result'])

```
