```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.impute import KNNImputer

# Step 1: Clean column - replace '?' and '\t?' with NaN
dataset["rc"] = dataset["rc"].replace(["?", "\t?"], np.nan)

# Step 2: Convert to numeric (in case column is still object dtype)
dataset["rc"] = pd.to_numeric(dataset["rc"], errors="coerce")

# Step 3: Scale features before imputation
scaler = StandardScaler()
data_scaled = scaler.fit_transform(dataset[["rc"]])

# Step 4: Apply KNN Imputer
imputer = KNNImputer(n_neighbors=3)
data_imputed = imputer.fit_transform(data_scaled)

# Step 5: Inverse scaling back to original scale
data_imputed = scaler.inverse_transform(data_imputed)

# Step 6: Update dataset
imputed_df = pd.DataFrame(data_imputed, columns=["rc"])
dataset["rc"] = np.round(imputed_df["rc"]).astype(int)

dataset.head()
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.impute import KNNImputer

# Step 1: Clean column - replace '?' and '\t?' with NaN
dataset["wc"] = dataset["wc"].replace(["?", "\t?"], np.nan)

# Step 2: Convert to numeric (in case column is still object dtype)
dataset["wc"] = pd.to_numeric(dataset["wc"], errors="coerce")

# Step 3: Scale features before imputation
scaler = StandardScaler()
data_scaled = scaler.fit_transform(dataset[["wc"]])

# Step 4: Apply KNN Imputer
imputer = KNNImputer(n_neighbors=3)
data_imputed = imputer.fit_transform(data_scaled)

# Step 5: Inverse scaling back to original scale
data_imputed = scaler.inverse_transform(data_imputed)

# Step 6: Update dataset
imputed_df = pd.DataFrame(data_imputed, columns=["wc"])
dataset["wc"] = np.round(imputed_df["wc"]).astype(int)

dataset.head()
```

```python
from sklearn.preprocessing import LabelEncoder

df_encoded = dataset.copy()

# Clean classification column specifically (remove tabs/spaces/newlines)
df_encoded["classification"] = df_encoded["classification"].str.strip()


```

```python

label_encoder = LabelEncoder()

# Encode all categorical columns
for col in df_encoded.select_dtypes(include=["object", "category"]).columns:
    df_encoded[col] = label_encoder.fit_transform(df_encoded[col])
```

```python
import pickle
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Features and target
X = df_encoded.drop(['id', 'classification'], axis=1)
y = df_encoded['classification']

# Initialize Linear Regression
estimator = LinearRegression()

# Apply RFE to select top 5 features
rfe = RFE(estimator=estimator, n_features_to_select=5)
rfe.fit(X, y)

# Get selected features
selected_features = X.columns[rfe.support_].tolist()
print("Selected Features:", selected_features)



```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
import pandas as pd

# Features and target
X = df_encoded.drop(['id', 'classification'], axis=1)
y = df_encoded['classification']

# Initialize Linear Regression
estimator = LinearRegression()

# Apply RFE to select top 5 features
rfe = RFE(estimator=estimator, n_features_to_select=5)
rfe.fit(X, y)

# Get selected features
selected_features = X.columns[rfe.support_].tolist()
print("Selected Features:", selected_features)

# Save selected features to CSV
pd.DataFrame(selected_features, columns=['Selected_Features']).to_csv('selected_features.csv', index=False)

print("Selected features saved to selected_features.csv successfully!")

```
