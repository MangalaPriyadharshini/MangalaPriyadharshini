```python
#importing the Libraies
import numpy as np
import pandas as pd
```

```python
# Reading the Dataset
dataset = pd.read_csv('/content/Social_Network_Ads.csv')
```

```python
dataset.head()
```

```python
dataset.tail()
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
X=dataset[['Age', 'EstimatedSalary','Gender_Male']]
y=dataset['Purchased']
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, cross_val_predict

# Define parameter grid for Logistic Regression
param_grid = {
    'penalty': ['l1', 'l2', 'elasticnet', 'none'],   # types of regularization
    'C': [0.01, 0.1, 1, 10, 100],                    # inverse of regularization strength
    'solver': ['saga', 'liblinear'],                 # solvers that support l1/elasticnet
    'l1_ratio': [None, 0.5]                          # only used when penalty='elasticnet'
}

# Initialize GridSearchCV with LogisticRegression
grid = GridSearchCV(
    estimator=LogisticRegression(max_iter=5000, random_state=42),
    param_grid=param_grid,
    refit=True,
    verbose=3,
    n_jobs=-1,
    scoring='f1_weighted'   # weighted F1 for imbalanced classes
)

# Fit the grid search
grid.fit(X, y)
```

```python
from sklearn.model_selection import cross_val_predict

# Use the best estimator from GridSearchCV
best_model = grid.best_estimator_

# Perform cross-validated predictions
y_pred = cross_val_predict(best_model, X, y, cv=5, n_jobs=-1)
```

```python
import pickle

# Save the best Logistic Regression model from GridSearchCV
filename = "LogisticRegression_CV_best_model.sav"
pickle.dump(grid.best_estimator_, open(filename, "wb"))

# Load the model
load_model = pickle.load(open("LogisticRegression_CV_best_model.sav", "rb"))

# Predict using the loaded model
y_pred = load_model.predict(X)
```
