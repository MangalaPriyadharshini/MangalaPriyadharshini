```python
#MBA ADMISSION DATASET
#Importing pandas
import pandas as pd
```

```python
#Load Data
dataset=pd.read_csv('/content/MBA.csv')
```

```python
#Printing dataset
dataset
```

```python

# Remove unwanted columns from the Dataset
X = dataset.drop(['international', 'race', 'admission'], axis=1)

# Target column
y = dataset['admission']

```

```python
#To perform One-Hot Encoding
X_enc = pd.get_dummies(X, drop_first=True, dtype=int)
```

```python
#printing
X_enc
```

```python
#Importing libraries for splitting the values
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_enc, y, test_size=0.2, random_state=42)
```

```python
#Importing Decision tree classifier
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
filename = "DecisionTreeClassifier.sav"
```

```python
import pickle
```

```python
#To perform accuracy
pickle.dump(model,open(filename,'wb'))
```
